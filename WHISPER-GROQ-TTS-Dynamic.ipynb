```python
##managing conversation history such after every 10 conversations we remove the first three least recently used
import numpy as np
from IPython.display import Audio
import gradio as gr
import soundfile as sf
import pdfplumber
import json
import requests
from groq import Client


client = Client(api_key='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')

target_dtype = np.int16
max_range = np.iinfo(target_dtype).max


conversation_history = []
document_content = ""

document_path = "/content/sample_18pg.pdf"

def read_document(document_path):
    global document_content


    with pdfplumber.open(document_path) as pdf:
        document_content = ""
        for page in pdf.pages:
            document_content += page.extract_text()

    print("Document Content:")
    print(document_content)

def manage_conversation_history():
    global conversation_history
    if len(conversation_history) > 20:
        conversation_history = conversation_history[6:]

def speech_to_speech_translation(audio_file):
    global conversation_history, document_content

    if not document_content:
        read_document(document_path)

    input_audio, sample_rate = sf.read(audio_file)

    response = transcribe(audio_file)
    translated_text = response

    conversation_history.append({"role": "user", "content": translated_text})

    print("Conversation History (before Groq response):")
    for msg in conversation_history:
        print(msg)

    initial_message = {
        "role": "system",
        "content": "Refer to the following document content for context: " + document_content
    }
    messages = [initial_message] + conversation_history

    chat_completion = client.chat.completions.create(
        messages=messages,
        model="llama3-8b-8192",
        max_tokens=100
    )
    groq_response = chat_completion.choices[0].message.content

    conversation_history.append({"role": "assistant", "content": groq_response})

    print("Conversation History (after Groq response):")
    for msg in conversation_history:
        print(msg)

    manage_conversation_history()

    synthesised_speech = generate_audio(groq_response)

    return synthesised_speech

read_document(document_path)

# Gradio interface
demo = gr.Blocks()

mic_translate = gr.Interface(
    fn=speech_to_speech_translation,
    inputs=gr.Audio(source="microphone", type="filepath"),
    outputs=gr.Audio(label="Generated Speech", type="numpy"),
)

file_translate = gr.Interface(
    fn=speech_to_speech_translation,
    inputs=gr.Audio(source="upload", type="filepath"),
    outputs=gr.Audio(label="Generated Speech", type="numpy"),
)

with demo:
    gr.TabbedInterface(
        [mic_translate, file_translate],
        ["Microphone", "Audio File"]
    )

demo.launch(debug=True)
```

Let's take a huge audio file it would divide the audio signal  to smaller chunks of this size 1 sec to 5sec and then these small chunks must be overlapping frames each of these audio signals are preprocessed and converted to mel spectrogram by fourier transform and the dimension of spectrogram would be F x T ,the mel spectrograms are also combined by stacking them along the time axis and we get a 3d tensor which  is fed into whisper model which is the shared encoder and decoder model and output text is generated.

and for the text to speech model it does not take each word rather the entire sentence and then convert them to phonemes and embeddings layers and then uses vocoder to convert to audio signal from the output of the lstm which includes spectrogram prediction
